```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf, explode
from pyspark.sql.types import ArrayType, StringType
import re

# Initialize Spark session
spark = SparkSession.builder.appName("MatchWords").getOrCreate()


# Define a UDF to find matched words
def find_matches(text, words):
    pattern = re.compile(r'\b(' + '|'.join(words) + r')\b', re.IGNORECASE)
    return pattern.findall(text)

# Register the UDF
find_matches_udf = udf(lambda text: find_matches(text, words_to_match), ArrayType(StringType()))

# Apply the UDF and extract matches
df_with_matches = df.withColumn("MatchedWords", find_matches_udf(col("Text")))

# Explode the matched words into separate rows if needed
df_exploded = df_with_matches.withColumn("MatchedWord", explode(col("MatchedWords"))).select("ID", "Text", "MatchedWord")

# Show the results
print("DataFrame with matched words:")
#df_with_matches.show(truncate=False)
display(df_with_matches)
print("Exploded DataFrame with each matched word in separate rows:")
#df_exploded.show(truncate=False)
```

```
from pyspark.sql.functions import size
# Filter out rows where 'values' column is an empty list
df_with_matches_FilterEmpty = df_with_matches.filter(size(df_with_matches["MatchedWords"]) > 0)

display(df_with_matches_FilterEmpty)
```

```
from pyspark.sql.functions import struct
#df_with_matches.dropDuplicates(["MatchedWords"])

# Remove duplicate rows
df_with_matches_dropDuplicates = df_with_matches.dropDuplicates("MatchedWords")
display(df_with_matches_dropDuplicates)
```

```
from pyspark.sql.functions import *
# Split the sentence into words and explode the list into rows
#word_counts = df_with_matches.select(explode(split(col("MatchedWords"), " ")).alias("word"))
word_counts = df_with_matches_dropDuplicates.select(explode(col("MatchedWords")).alias("word"))
# Group by the word and count occurrences
word_counts_result = word_counts.groupBy("word").count()

display (word_counts_result)
```
